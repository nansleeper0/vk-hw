## üß© –ó–∞–¥–∞—á–∞ ‚Ññ1. –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –∫—É—á–∏

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä–∏—Ç –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –∫—É—á–∏

```python
def isMaxHeap(arr):
    n = len(arr)
    for i in range((n - 2) // 2 + 1):
        left = 2 * i + 1
        right = 2 * i + 2

        if left < n and arr[i] < arr[left]:
            return False

        if right < n and arr[i] < arr[right]:
            return False

    return True
```

```python
from collections import deque

def isMaxHeap(arr):
    n = len(arr)
    if n <= 1:
        return True

    queue = deque()
    queue.append(0)  

    while queue:
        i = queue.popleft()

        left = 2 * i + 1
        right = 2 * i + 2

        if left < n:
            if arr[i] < arr[left]:
                return False
            queue.append(left)

        if right < n:
            if arr[i] < arr[right]:
                return False
            queue.append(right)

    return True
```

```python
from collections import deque

def isMaxHeap(root):
    if root is None:
        return True

    queue = deque([root])
    should_be_leaf = False

    while queue:
        current = queue.popleft()

        if current.left:
            if should_be_leaf or current.left.val > current.val:
                return False
            queue.append(current.left)
        else:
            should_be_leaf = True

        if current.right:
            if should_be_leaf or current.right.val > current.val:
                return False
            queue.append(current.right)
        else:
            should_be_leaf = True

    return True
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ2. –ü–æ–ª–Ω–æ–µ –±–∏–Ω–∞—Ä–Ω–æ–µ –¥–µ—Ä–µ–≤–æ

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ
–≠—Ç–æ –¥–µ—Ä–µ–≤–æ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É—Ä–æ–≤–Ω–∏, –∫—Ä–æ–º–µ
–ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ, –ø–æ–ª–Ω–æ—Å—Ç—å—é –∑–∞–ø–æ–ª–Ω–µ–Ω—ã, –∞
—É–∑–ª—ã –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —É—Ä–æ–≤–Ω—è –∑–∞–ø–æ–ª–Ω—è—é—Ç—Å—è
—Å–ª–µ–≤–∞ –Ω–∞–ø—Ä–∞–≤–æ.
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è
–ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –¥–∞–Ω–Ω–æ–µ –±–∏–Ω–∞—Ä–Ω–æ–µ
–¥–µ—Ä–µ–≤–æ –ø–æ–ª–Ω—ã–º.

---


```python
from collections import deque

def isCompleteTree(root):
    if not root:
        return True

    queue = deque([root])
    should_be_leaf = False

    while queue:
        node = queue.popleft()

        if node.left:
            if should_be_leaf:
                return False
            queue.append(node.left)
        else:
            should_be_leaf = True

        if node.right:
            if should_be_leaf:
                return False
            queue.append(node.right)
        else:
            should_be_leaf = True

    return True
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ3. –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ K –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –º–∞—Å—Å–∏–≤–æ–≤

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ
–†–µ–∞–ª–∏–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç K
–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –º–∞—Å—Å–∏–≤–æ–≤ –≤ –æ–¥–∏–Ω
–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –º–∞—Å—Å–∏–≤. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –º–∏–Ω-
–∫—É—á—É –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
—Ç–µ–∫—É—â–∏—Ö –º–∞—Å—Å–∏–≤–æ–≤, —á—Ç–æ –ø–æ–∑–≤–æ–ª–∏—Ç –∏–∑–≤–ª–µ–∫–∞—Ç—å –∏—Ö
–ø–æ –æ—á–µ—Ä–µ–¥–∏, —Å–æ—Ö—Ä–∞–Ω—è—è –ø–æ—Ä—è–¥–æ–∫.

---


```python
import heapq

def merge_k_sorted_arrays(sorted_arrays):
    min_heap = []

    for arr in sorted_arrays:
        for x in arr:
            heapq.heappush(min_heap, x)

    merged = []

    while min_heap:
        merged.append(heapq.heappop(min_heap))

    return merged
```

```python
import heapq

def merge_k_sorted_arrays(sorted_arrays):
    min_heap = []
    merged = []

    for array_idx, arr in enumerate(sorted_arrays):
        if len(arr) > 0:
            heapq.heappush(min_heap, (arr[0], array_idx, 0))

    while min_heap:
        value, array_idx, element_idx = heapq.heappop(min_heap)
        merged.append(value)

        next_idx = element_idx + 1
        if next_idx < len(sorted_arrays[array_idx]):
            next_value = sorted_arrays[array_idx][next_idx]
            heapq.heappush(min_heap, (next_value, array_idx, next_idx))

    return merged
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ4. –ö-—ã–π –Ω–∞–∏–º–µ–Ω—å—à–∏–π/–Ω–∞–∏–±–æ–ª—å—à–∏–π —ç–ª–µ–º–µ–Ω—Ç –≤ BST

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–ù–∞–π—Ç–∏ –∫-—ã–π –Ω–∞–∏–º–µ–Ω—å—à–∏–π –∏ –Ω–∞–∏–±–æ–ª—å—à–∏–π —ç–ª–µ–º–µ–Ω—Ç

---



```python
def inorder_min_iterative(root, k):
    stack = []
    current = root
    counter = 0

    while stack or current:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        counter += 1

        if counter == k:
            return current.val

        current = current.right

    return None  
```

```python
def inorder_min(node, k, counter):
    if node is None:
        return None

    left_result = inorder_min(node.left, k, counter)
    if left_result is not None:
        return left_result

    counter[0] += 1

    if counter[0] == k:
        return node.val

    return inorder_min(node.right, k, counter)
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ5. Balance factor

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–ë–∞–ª–∞–Ω—Å-—Ñ–∞–∫—Ç–æ—Ä —É–∑–ª–∞ –≤ –¥–µ—Ä–µ–≤–µ
—Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ —Ä–∞–∑–Ω–∏—Ü–∞ –≤—ã—Å–æ—Ç –ª–µ–≤–æ–≥–æ
–∏ –ø—Ä–∞–≤–æ–≥–æ –ø–æ–¥–¥–µ—Ä–µ–≤—å–µ–≤:

---




```python
def calculate_heights_and_balance(node):
    if node is None:
        return 0  
        
    left_height = calculate_heights_and_balance(node.left)
    right_height = calculate_heights_and_balance(node.right)

    node.balanceFactor = left_height - right_height

    return 1 + max(left_height, right_height)
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ6. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ –∑–µ—Ä–∫–∞–ª—å–Ω–æ–µ –¥–µ—Ä–µ–≤–æ

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–ù—É–∂–Ω–æ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º,
–∫–æ—Ç–æ—Ä—ã–π –ø–µ—Ä–µ–≤–µ—Ä–Ω–µ—Ç –±–∏–Ω–∞—Ä–Ω–æ–µ
–¥–µ—Ä–µ–≤–æ "–≤–≤–µ—Ä—Ö –Ω–æ–≥–∞–º–∏", —Ç.–µ.
–ø–æ–º–µ–Ω—è–µ—Ç –º–µ—Å—Ç–∞–º–∏ –ª–µ–≤—ã–µ –∏ –ø—Ä–∞–≤—ã–µ
–ø–æ–¥–¥–µ—Ä–µ–≤—å—è –∫–∞–∂–¥–æ–≥–æ —É–∑–ª–∞.

---




```python
def mirror_tree(node):
    if node is None:
        return None

    node.left, node.right = node.right, node.left

    mirror_tree(node.left)
    mirror_tree(node.right)

    return node
```

```python
from collections import deque

def mirror_tree_iterative(node):
    if node is None:
        return None

    queue = deque([node])

    while queue:
        current = queue.popleft()

        current.left, current.right = current.right, current.left

        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

    return node
```
