### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π –¥–ª–∏–Ω—ã N
—Å–æ—Å—Ç–æ—è—â–µ–π –∏–∑ 0 –∏ 1 –≤ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ
–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π –∏–∑ –Ω—É–ª–µ–π –∏
–µ–¥–∏–Ω–∏—Ü –¥–ª–∏–Ω—ã, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–∏–∫–∞–∫–∏–µ
—Ç—Ä–∏ –µ–¥–∏–Ω–∏—Ü—ã –Ω–µ —Å—Ç–æ—è—Ç —Ä—è–¥–æ–º.

---


```python
def count_sequences(n):
    if n == 0:
        return 1
    if n == 1:
        return 2
    if n == 2:
        return 4
    dp = [1, 2, 4]
    for i in range(3, n + 1):
        dp.append(dp[i - 1] + dp[i - 2] + dp[i - 3])
    return dp[n]
```

### üìò –û–ø–∏—Å–∞–Ω–∏–µ
–î–∞–Ω –º–∞—Å—Å–∏–≤ –Ω–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —á–∏—Å–µ–ª. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–ª–∏–Ω–Ω—É—é
–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—É—é –≤–æ–∑—Ä–∞—Å—Ç–∞—é—â—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏ –≤–µ—Ä–Ω—É—Ç—å –µ–µ –¥–ª–∏–Ω—É

[3, 2, 8, 9, 5, 10] - –æ—Ç–≤–µ—Ç 3 —Ç–∞–∫ –∫–∞–∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–ª–∏–Ω–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å
2, 8, 9

[1, 2, 7, 9, 0, 10] - –æ—Ç–≤–µ—Ç 4 —Ç–∞–∫ –∫–∞–∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–ª–∏–Ω–Ω–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å
1, 2, 7, 9

[8, 8, 8, 8] - –æ—Ç–≤–µ—Ç 1

---


```python
def findLIS(nums):
    if len(nums) == 0:
        return 0
    if len(nums) == 1:
        return 1
    dp = [1] * len(nums)
    for i in range(1, len(nums)):
        if nums[i - 1] < nums[i]:
            dp[i] = dp[i - 1] + 1
    return max(dp)
```

```python
def pascal_triangle(row, col):
    if col == 0 or row == col:
        return 1
    return pascal_triangle(row - 1, col - 1) + pascal_triangle(row - 1, col)
```

```python
n = 6

dp = []
for i in range(1, n + 1):
    tmp = []
    for j in range(1, i + 1):
        tmp.append(1)
    dp.append(tmp)

for row in range(1, n):
    for col in range(1, row):
        dp[row][col] = dp[row - 1][col - 1] + dp[row - 1][col]
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ5. –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –≤—ã–≥–æ–¥–∞

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–î–∞–Ω –º–∞—Å—Å–∏–≤ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª. –ö–∞–∂–¥–æ–µ
—á–∏—Å–ª–æ - —Å—Ç–æ–∏–º–æ—Å—Ç—å –∞–∫—Ü–∏–∏. –ù–∞–º –Ω—É–∂–Ω–æ
–∫—É–ø–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–µ—à–µ–≤–æ, –∞
–ø—Ä–æ–¥–∞—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –¥–æ—Ä–æ–≥–æ.
–°–¥–µ–ª–∞—Ç—å —ç—Ç–æ –Ω–∞–¥–æ –∑–∞ O(n)

---




```python
def maxProfit(prices):
    profit = 0
    min_price = prices[0]
    for i in range(1, len(prices)):
        profit = max(profit, prices[i] - min_price)
        min_price = min(prices[i], min_price)
    return profit
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ6. –†–∞–∑–º–µ–Ω –º–æ–Ω–µ—Ç

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–î–∞–Ω –º–∞—Å—Å–∏–≤ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª coin, –≥–¥–µ –∫–∞–∂–¥–æ–µ
—á–∏—Å–ª–æ - –Ω–æ–º–∏–Ω–∞–ª –º–æ–Ω–µ—Ç—ã –∏ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ
amount - —Å—É–º–º–∞ –º–æ–Ω–µ—Ç –∏–∑ –º–∞—Å—Å–∏–≤–∞.
–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ
–º–æ–Ω–µ—Ç, –∫–æ—Ç–æ—Ä–æ–µ –≤ —Å—É–º–º–µ –¥–∞–ª–æ –±—ã amount. –ï—Å–ª–∏
—Ç–∞–∫–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –Ω–µ—Ç - –≤–æ–∑–≤—Ä–∞—â–∞–µ–º -1

---




```python
import sys

def coinChange(coins, amount):
    if amount == 0:
        return 0
    if amount < 0:
        return -1

    min_coins = sys.maxsize

    for coin in coins:
        res = coinChange(coins, amount - coin)
        if res >= 0 and res < min_coins:
            min_coins = res + 1

    if min_coins == sys.maxsize:
        return -1

    return min_coins
```

```python
def coin_change(coins, amount, cache=None):
    if cache is None:
        cache = {}

    if amount == 0:
        return 0
    if amount < 0:
        return -1
    if amount in cache:
        return cache[amount]

    min_coins = sys.maxsize

    for coin in coins:
        res = coin_change(coins, amount - coin, cache)
        if res >= 0 and res < min_coins:
            min_coins = res + 1

    if min_coins == sys.maxsize:
        cache[amount] = -1
    else:
        cache[amount] = min_coins

    return cache[amount]
```

```python
def coinChange(coins, amount):
    dp = [sys.maxsize] * (amount + 1)
    dp[0] = 0

    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    if dp[amount] == sys.maxsize:
        return -1
    return dp[amount]
```

## üß© –ó–∞–¥–∞—á–∞ ‚Ññ7. –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–∞–ª–∏–Ω–¥—Ä–æ–º –≤ —Å—Ç—Ä–æ–∫–µ

---

### üìò –û–ø–∏—Å–∞–Ω–∏–µ

–î–∞–Ω–∞ —Å—Ç—Ä–æ–∫–∞ s. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—É—é –ø–æ–¥—Å—Ç—Ä–æ–∫—É, –∫–æ—Ç–æ—Ä–∞—è —è–≤–ª—è–µ—Ç—Å—è
–ø–∞–ª–∏–Ω–¥—Ä–æ–º–æ–º

---




```python
def longestPalindrome(s):
    currentMaxLeft = 0
    currentMaxRight = 0

    def expandAroundCenter(l, r):
        nonlocal currentMaxLeft, currentMaxRight
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if (r - l) > (currentMaxRight - currentMaxLeft):
                currentMaxLeft = l
                currentMaxRight = r
            l -= 1
            r += 1

    for i in range(len(s)):
        expandAroundCenter(i, i)
        expandAroundCenter(i, i + 1)

    return s[currentMaxLeft:currentMaxRight + 1]
```

```python
def longestPalindrome(s):
    n = len(s)
    if n == 0:
        return ""

    dp = [[False] * n for _ in range(n)]

    start = 0
    max_len = 1

    for i in range(n):
        dp[i][i] = True

    for i in range(n - 1):
        if s[i] == s[i + 1]:
            dp[i][i + 1] = True
            start = i
            max_len = 2

    for length in range(3, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1
            if s[i] == s[j] and dp[i + 1][j - 1]:
                dp[i][j] = True
                start = i
                max_len = length

    return s[start:start + max_len]
```
